## 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt

with open('datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = [] # 연령별 인구 저장할 리스트


    for row in data:
        if '상대원3동' in row[0]:
            for i in row[3:]:
                result.append(int(i.replace(",","")))
    print(len(row[3:]))
    print(result)
    plt.rc('font', family='Malgun Gothic')
    plt.style.use("ggplot")
    #plt.plot(result)
    plt.bar(result)
    plt.xlabel('연령')
    plt.ylabel('인구')
    plt.show()
```
